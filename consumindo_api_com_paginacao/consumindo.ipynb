# consumindo com paginação e amarzemando em dataframe e amazenando em lista

```python
import pandas as pd
import requests


page = 1


dados_coletados = []


params = {
    "_page": page,
    "_limit": 10 
}

while True:
    url = "https://jsonplaceholder.typicode.com/posts"
    params["_page"] = page
    resposta = requests.get(url, params=params)
    
    if resposta.status_code != 200:
        print(f"Erro ao acessar a API: {resposta.status_code}")
        break
    informacoes = resposta.json()
    
    if not informacoes:
        break
    
    dados_coletados.extend(informacoes)
    
    page += 1

# Converte a lista de dados em um DataFrame
nova_tabela = pd.DataFrame(dados_coletados)

# Salva o DataFrame final em um arquivo Excel
excel_file = "dados_coletados.xlsx"
nova_tabela.to_excel(excel_file, index=False)

print(f"Dados salvos em: {excel_file}")
```

# consumindo com paginação e amarzemando em dataframe e amazenando em lista e adicionando limite de requisição por tempo

```python
import pandas as pd
import requests
import time

# Função para coletar dados de uma página específica
def coletar_dados(page):
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {
        "_page": page,
        "_limit": 10  # Limite de registros por página
    }
    resposta = requests.get(url, params=params)
    return resposta

# Inicialização
page = 1
nova_tabela = pd.DataFrame()
limite_por_pagina = 10
tempo_limite_minutos = 15

# Contadores e controle de tempo
requisicoes_feitas = 0
tempo_inicio = time.time()

while True:
    # Verifica se já atingiu o limite de requisições por página
    if requisicoes_feitas >= limite_por_pagina:
        # Calcula o tempo decorrido desde o último reset
        tempo_decorrido = time.time() - tempo_inicio
        # Se o tempo decorrido for inferior a 15 minutos, espera o tempo restante
        if tempo_decorrido < (tempo_limite_minutos * 60):
            tempo_espera = (tempo_limite_minutos * 60) - tempo_decorrido
            print(f"Esperando {tempo_espera:.1f} segundos para continuar...")
            time.sleep(tempo_espera)
        # Reinicia o contador de requisições e tempo de início
        requisicoes_feitas = 0
        tempo_inicio = time.time()

    # Coleta os dados da página atual
    resposta = coletar_dados(page)

    # Verifica o código de status da resposta
    if resposta.status_code != 200:
        print(f"Erro ao acessar a API: {resposta.status_code}")
        break

    informacoes = resposta.json()
    if not informacoes:
        break
    tabela = pd.DataFrame(informacoes)
    nova_tabela = pd.concat([nova_tabela, tabela], ignore_index=True)
    page += 1
    requisicoes_feitas += 1

# Salva o DataFrame final em um arquivo Excel
excel_file = "dados_coletados.xlsx"
nova_tabela.to_excel(excel_file, index=False)

print(f"Dados salvos em: {excel_file}")
```

# Consumindo api com paginacao, colocando periodo e tipo de solicitacao

```python
import pandas as pd
import requests
import time
from datetime import datetime, timedelta

# Função para coletar dados de uma página específica com parâmetros de data
def coletar_dados(page, data_inicial, data_final, situacao):
    url = "https://sua_api.com/dados"
    params = {
        "_page": page,
        "data_inicial": data_inicial.strftime('%d-%m-%Y'),
        "data_final": data_final.strftime('%d-%m-%Y'),
        "situacao": situacao
    }
    resposta = requests.get(url, params=params)
    return resposta

# Inicialização
limite_por_pagina = 10
tempo_limite_minutos = 15
requisicoes_feitas = 0
tempo_inicio = time.time()

# Define os intervalos de consulta para cada tipo de situacao
intervalo_dias = 180
hoje = datetime.now().date()

# Para nova, calcula as datas
data_nova_inicial = hoje - timedelta(days=60)
data_nova_final = hoje + timedelta(days=60)

# Para em_andamento e finalizada, defina os intervalos de datas conforme necessário
data_em_andamento_inicial = hoje - timedelta(days=90)
data_em_andamento_final = hoje - timedelta(days=30)

data_finalizada_inicial = hoje - timedelta(days=180)
data_finalizada_final = hoje - timedelta(days=91)

# Lista de configurações de consulta por situacao
configuracoes_consulta = [
    {"situacao": "nova", "data_inicial": data_nova_inicial, "data_final": data_nova_final},
    {"situacao": "em_andamento", "data_inicial": data_em_andamento_inicial, "data_final": data_em_andamento_final},
    {"situacao": "finalizada", "data_inicial": data_finalizada_inicial, "data_final": data_finalizada_final}
]

# Loop principal para cada configuração de consulta
for config in configuracoes_consulta:
    situacao = config["situacao"]
    data_inicial = config["data_inicial"]
    data_final = config["data_final"]
    
    page = 1
    nova_tabela = pd.DataFrame()
    
    while True:
        # Verifica se já atingiu o limite de requisições por página
        if requisicoes_feitas >= limite_por_pagina:
            # Calcula o tempo decorrido desde o último reset
            tempo_decorrido = time.time() - tempo_inicio
            # Se o tempo decorrido for inferior a 15 minutos, espera o tempo restante
            if tempo_decorrido < (tempo_limite_minutos * 60):
                tempo_espera = (tempo_limite_minutos * 60) - tempo_decorrido
                print(f"Esperando {tempo_espera:.1f} segundos para continuar...")
                time.sleep(tempo_espera)
            # Reinicia o contador de requisições e tempo de início
            requisicoes_feitas = 0
            tempo_inicio = time.time()
        
        # Coleta os dados da página atual com os parâmetros de data e situacao
        resposta = coletar_dados(page, data_inicial, data_final, situacao)
        
        # Verifica o código de status da resposta
        if resposta.status_code != 200:
            print(f"Erro ao acessar a API: {resposta.status_code}")
            break
        
        informacoes = resposta.json()
        if not informacoes:
            break
        
        tabela = pd.DataFrame(informacoes)
        nova_tabela = pd.concat([nova_tabela, tabela], ignore_index=True)
        page += 1
        requisicoes_feitas += 1
    
    # Salva o DataFrame final em um arquivo Excel com nome baseado na situacao
    excel_file = f"dados_coletados_{situacao}.xlsx"
    nova_tabela.to_excel(excel_file, index=False)
    
    print(f"Dados da situacao '{situacao}' salvos em: {excel_file}")

print("Processo concluído.")
```
