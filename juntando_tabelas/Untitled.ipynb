# merge

```python
tabela_produtos = pd.merge(tabela_python, tabela_cienciados, on="Cliente")
display(tabela_produtos)
```

```python
tabela_produtos = pd.merge(tabela_python, tabela_cienciados, left_on="Cliente", right_on="Cliente")
display(tabela_produtos)
```

```python
# analisando o parametro how
# inner => Inclui as informações que sõ possui nas duas tabelas
# outher => Mostra tudo
# left
# right
tabela_produtos = pd.merge(tabela_python, tabela_cienciados, on="Cliente", how="left")
display(tabela_produtos)
```

```python
# analisando o parametro how
# inner
# outher
# left
# right
tabela_produtos = pd.merge(tabela_python, tabela_cienciados, on="Cliente", how="outer")
display(tabela_produtos)
```
