

# Organização e tratamento dos dados

## nomeando variaveis

```python
covid_sp.head(3)
```

```python
# renomeando coluna
covid_sp = covid_sp.rename(columns={'nome_munic': 'municipio'})
```

```python
covid_sp.head(3)
```

```python
# Outra forma de alterar nome de coluna
# OBS: inplace=True é para salvar a alteração no dataFrame
covid_sp.rename(columns={'datahora': 'data'}, inplace=True)
covid_sp.head(3)
```

```python
# Renomear mais de uma coluna
covid_sp.rename(columns={'map_leg':'rotulo_mapa','map_leg_s':'codigo_mapa'}, inplace=True)
covid_sp.head(3)
```

```python
# Excluir coluna por nome

covid_sp_alterado = covid_sp.drop(columns=['cod_ra'])
covid_sp_alterado.shape
```

```python
# Excluir por numero, pelo indice da coluna
covid_sp_alterado = covid_sp_alterado.drop(covid_sp_alterado.columns[[1]],axis=1)
covid_sp_alterado.shape
```

```python
# Excluir mais de uma coluna por nome
covid_sp_alterado.drop(columns=['rotulo_mapa','codigo_mapa','cod_drs'],inplace=True)
covid_sp_alterado.head(3)
```

```python
# Excluir varias colunas pelo indice da coluna
covid_sp_alterado.drop(covid_sp_alterado.columns[[13,14,13,19]], axis=1, inplace=True)
covid_sp_alterado.shape
```



# Criando e alterando valores das colunas

```python
covid_sp_alterado.head(3)
```

```python
# Verificamos que a coluna "area" esta 100x maior que o normal, 
# EX: ao invez de retornar 411.781 esta retornando 41199
covid_sp_alterado['area'] = covid_sp_alterado['area']/100
# pode usar também
# covid_sp_alterado['area'] = covid_sp_alterado.area/100
```

```python
# Criação de uma coluna com a densidade demográfica (hab/km2)
covid_sp_alterado['densidade'] = covid_sp_alterado['pop'] / covid_sp_alterado['area']
```

```python
covid_sp_alterado.shape
```

```python
# Criação da coluna com os indices, para começar a partir do 1
lista = list(range(1,374035))
#lista
```

```python
#transforma lista em dataframe
df = pd.DataFrame(lista,columns=['indice'])
df
```

```python
# Juntando os dois DataFrame
covid_sp_alterado = pd.concat([covid_sp_alterado,df],axis=1)
covid_sp_alterado.head()
```
